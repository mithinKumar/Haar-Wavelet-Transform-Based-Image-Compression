```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Function to display an image
def display_image(img, title="Image"):
    plt.imshow(img, cmap='gray')
    plt.title(title)
    plt.axis('off')
    plt.show()

# Load the image
image_path = '/content/testcat.jpg'  # Change this to your image path
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Resize the image to 64x64
resized_image = cv2.resize(image, (64, 64), interpolation=cv2.INTER_AREA)

# Display the resized image
#display_image(resized_image, title="Resized 64x64 Image")

# Convert the image pixels to 8-bit hexadecimal and print the data statements
data_statements = []
for i in range(64):
    for j in range(64):
        pixel_value = resized_image[i, j]
        hex_value = format(pixel_value, '02x')  # Convert to 2-digit hexadecimal
        index = i * 64 + j
        data_statements.append(f'data({index}) <= x"{hex_value}";')

# Print the data statements
for statement in data_statements:
    print(statement)
```

```python
# Read the input file and generate a new output file
input_file = 'haar.txt'  # Replace with your input file name
output_file = 'output.txt'  # Output file name

with open(input_file, 'r') as infile, open(output_file, 'w') as outfile:
    for line in infile:
        columns = line.split()
        # Keep only the 3rd to 6th columns (index 2 to 5)
        new_columns = columns[2:6]
        # Join them back into a string and write to the output file
        outfile.write(' '.join(new_columns) + '\n')

print(f"Processed file saved as {output_file}.")
```

```python
# Read the input file and generate a new output file with hex values
input_file = 'output.txt'  # Replace with your input file name
output_file = 'output_hex.txt'  # Output file name

with open(input_file, 'r') as infile, open(output_file, 'w') as outfile:
    for line in infile:
        # Split the line into columns (8-bit binary numbers)
        binary_numbers = line.split()
        # Convert each binary number to hex and format it
        hex_numbers = [format(int(b, 2), '02X') for b in binary_numbers]
        # Join the hex numbers into a string and write to the output file
        outfile.write(' '.join(hex_numbers) + '\n')

print(f"Processed file saved as {output_file}.")
```

```python
# Read the input file and generate a new output file with even rows
input_file = 'output_hex.txt'  # Replace with your input file name
output_file = 'even_rows.txt'  # Output file name

with open(input_file, 'r') as infile, open(output_file, 'w') as outfile:
    for index, line in enumerate(infile):
        # Check if the row index is even (0-based index)
        if index % 2 == 0:
            outfile.write(line)

print(f"Processed file saved as {output_file}.")
```

```python
# Read the input file and store the data
input_file = 'even_rows.txt'  # Replace with your input file name
output_file = 'outputf.txt'  # The file where you want to save the output

# Initialize an empty list to hold the rows
data = []

# Read the data from the input file
with open(input_file, 'r') as f:
    for line in f:
        # Split the line into hex values and strip whitespace
        hex_values = line.strip().split()
        data.append(hex_values)

# Convert the 1024x4 data to 64x64
output_data = []
for i in range(64):
    new_row = []
    for j in range(64):
        # Calculate the original index in the 1024x4 structure
        original_index = i * 16 + (j // 4)  # 16 rows in the original data
        original_col = j % 4
        new_row.append(data[original_index][original_col])
    output_data.append(new_row)

# Write the output data to a new file
with open(output_file, 'w') as f:
    for row in output_data:
        f.write(' '.join(row) + '\n')

print("Transformation complete. Output saved to", output_file)
```
